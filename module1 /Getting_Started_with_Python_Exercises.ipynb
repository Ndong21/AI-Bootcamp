## Exercise 6: More Concepts on Python
Task: Write a program that calculates the factorial of a given number using a while loop.

```python
# Your code here

"""
Algorithm:
  1-get a number
  2-initialize factorial to 1
  3-do this steps repeated until number=1
    a- multiply number by factorial
    b- decrease number by 1
  4-print the factorial
"""
number = 5 #number to calculate the factorial
factorial = 1 # the results will be stored here

while number > 1: # check that number is greater than 1
  factorial = factorial * number # multiply number with what is stored in factorial

  number -=1 # decrease number by 1

print(factorial) #print the factorial of number
```

```python
# Your code here

#import math module

import math

# Ask user to enter a number
number = int(input("Enter a number to determine if its prime: "))

if number < 2:
  print("Not prime number")

else:
  """
  #Logic
  ---------
  for any divisor d of n , there is a corresponding divisor n/d .
  If d is less than or equal to the √n , then n/d will be greater than or
  equal to the √n . Therefore, we only need to check for divisors up to the √n.
  If n has a divisor larger than its square root, the paired divisor must be
  smaller than the square root, which we would have already checked.
  """
  for i in range(2, int(math.sqrt(number)) + 1):
    if number % i == 0:
      print("Not prime")
      break
  else:
    print("Its prime")


# This second code will do the check and return prime or not prime along with its factors

# number = int(input("Enter a number to determine if its prime: "))
# factors = []
# for i in range(1,number+1):

#   if number % i == 0:
#     factors.append(i)

# if len(factors) == 2:
#   print(f"Hurray!, {number} its prime guys with factors: {factors}")
# else:
#   print(f"It's not prime. {number} has the following factors: {factors}")
```

## Exercise 15: Fibonacci Sequence
Task: Write a program that prints the first 10 numbers in the Fibonacci sequence.

```python
# Your code here

# logic
"""
The next term is found by adding the two numbers before it
"""

def fib_sequence(n):
  sequence = []
  for i in range(0, n):
    if i == 0 or i == 1:
      sequence.append(i)
    else:
      next = sequence[-1] + sequence[-2]
      sequence.append(next)

  return sequence

fib_sequence(10)
```
