Password Generator: Write a program that generates a random password for the user. Ask the user how long they want their password and how many letters and numbers they want in their password. Have a mix of upper and lowercase letters, numbers, and symbols. The password should be a minimum of 6 characters long


```python
'''
Algorithm:
--------
  - ask users to enter required len and how many numbers, letters and symbols they want
  - randomly generate the numbers, letters, and symbols
  - combine the generated output and shuffle them
  - give out the result.

Checks
-------
 - check that the min password length entered by user is 6 or higher
 - check that the number of letters, numbers and symbols equal to the to the length specified.
'''
import random
import string

# ask users for password length
print("password must be atleast 6 characters")
length = int(input("Enter your desire password length: "))

while length < 6:
  print("password length too short, it must be atleast 6 characters")
  length = int(input("Enter your desire password length: "))

#ask users to enter password requirements
def get_requirements():
  print("Note!!, each password must contain atleast one number, letter or special symbol")
  numbers = int(input("Enter how many numbers you want in your password: "))
  letters = int(input("Enter how many letters you want in the password: "))
  symbols = int(input("Enter how many symbols you want in the password: "))

  return numbers, letters, symbols

# get initial requirements
numbers, letters, symbols = get_requirements()

#check that requirements are met
while length != numbers + letters +symbols:
  print("The total number of characters (numbers + letters + symbols) must be equal to the specified password length.")

  if length < numbers + letters +symbols:
    print("You have specified more characters than the password length.")

  else:
    print("You have specified fewer characters than the password length.")

  numbers, letters, symbols = get_requirements()

# functions to randomly generate numbers, letters and symbols

def get_random_letters(count):
  if count == 1:
    # ensure at least one letter is returned
    return random.choices(string.ascii_letters)

  # ensure at least one uppercase and one lowercase letter
  uppercase_letter = random.choice(string.ascii_uppercase)
  lowercase_letter = random.choice(string.ascii_lowercase)

  # fill the remaining letters randomly
  remaining_letters = random.choices(string.ascii_letters, k=count - 2)

  #combine and shuffle to mix cases
  all_letters = [uppercase_letter, lowercase_letter] + remaining_letters
  random.shuffle(all_letters)

  return "".join(all_letters)

def get_random_numbers(count):
  return "".join(random.choices(string.digits, k=count))

def get_random_symbols(count):
  return "".join(random.choices(string.punctuation, k=count))

# get user inputs for password specification

random_letters = get_random_letters(letters)
random_numbers = get_random_numbers(numbers)
random_symbols = get_random_symbols(symbols)

# generate a random password based on user requirements

combine_characters = random_letters + random_numbers + random_symbols
character_list = list(combine_characters)
random.shuffle(character_list)

password = "".join(character_list)
print(password)

```
